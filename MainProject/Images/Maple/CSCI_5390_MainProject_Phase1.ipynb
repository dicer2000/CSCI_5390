```python
# Check it out
import pandas as pd
df = pd.read_csv('TreeImageMaster.csv')
df.head()
```

```python
countOfTrees = df[['treetype', 'id']].groupby('treetype').count()
countOfTrees

# Legend
# Species       ID
#Ash           3
#Cedar         4
#Honeysuckle   0
#Maple         2
#Oak           1

```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(figsize=(7,5), dpi=100)
plt.bar(countOfTrees.index, height=countOfTrees.id)
plt.title('Distribution of Tree Species');
```
